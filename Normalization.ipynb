```python
import pandas as pd
import datetime
```

```python
trips = pd.read_excel('realdata.xlsx', sheet_name='trips')
towns = pd.read_excel('realdata.xlsx', sheet_name='towns')
```

```python
trips.head()
```

```python
pickup_routes = pd.DataFrame(trips['pickup'].unique()).reset_index()
pickup_routes.columns = ['ID_Pickup', 'pickup']
dropoff_routes = pd.DataFrame(trips['dropoff'].unique()).reset_index()
dropoff_routes.columns = ['ID_Dropoff', 'dropoff']
```

```python
#Merge
trips_1 = trips.merge(pickup_routes, how='left', left_on='pickup', right_on = 'pickup').\
                merge(dropoff_routes, how='left', left_on='dropoff', right_on = 'dropoff')
trips_1['Routes'] = trips_1['ID_Pickup'].astype('str') + '-' + trips_1['ID_Dropoff'].astype('str')
trips_1.head()
```

```python
#Trip Duration
trips_1['Trip Duration'] = trips_1['Dropoff Timestamp'] - trips_1['Pickup Timestamp']
trips_1['Trip Duration'] = trips_1['Trip Duration'].astype('timedelta64[m]')
```

```python
trips_1.head()
```

```python
trips_1.to_excel('real_data_1.xlsx', sheet_name='trip')
```
